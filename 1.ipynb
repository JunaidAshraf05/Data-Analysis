.head()

.shape

```python
data.shape
```

.index

```python
data.index
```

.columns

it shows the name of each columns


```python
data.columns
```

.dtypes

it shows the data-type of each columns

```python
data.dtypes
```

.unique()

in a column it shows all the unique values it can be applied on a single column only not on the whole database

```python
data["Weather"].unique()
```

.nunique()

it shows the total no. of unique in each  column it can be applied on a single column as well as onn the whole database

```python
data.nunique()
```

.count

it shows the total no of null values with their count it can be applied on single column only

```python
data.count()
```

.value_counts

in a column it shows all the unique values with their count it can be applied

```python
data['Weather'].value_counts()
```

.info()
